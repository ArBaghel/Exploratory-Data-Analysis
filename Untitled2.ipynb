```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/loan_test.csv')
df
```

```python
df.info()
```

```python
(df.isnull().sum()/df.shape[0])*100
```

```python
df["Gender"].value_counts()
```

```python
df["Gender"].isnull().sum()
```

```python
df["Gender"].mode()[0]
```

```python
df["Gender"].fillna(df["Gender"].mode()[0],inplace=True)
df
```

```python
df.dtypes
```

```python
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Dependents'].isnull().sum()
df
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
df['Credit_History'].fillna(df['Credit_History'].mode()[0],inplace=True)
df['Credit_History'].isnull().sum()
```

```python
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
df['Self_Employed'].isnull().sum()
```

```python
df['Term'].isnull().sum()
```

```python
df['Term'].fillna(method="bfill",inplace=True)
df['Term'].isnull().sum()
df
```

```python
df['Education'].replace({'Not Graduate':0,'Graduate':1})
df
```

```python
#Education
plt.hist(x=df['Education'])#.value_counts(),labels=["Graduate","Not Graduate"],autopct="%0.2f%%")
plt.title("Education")
plt.show()
plt.pie(x=df['Dependents'].value_counts(),labels=["0","1","2","3+"],autopct="%0.2f%%")
plt.title("Dependents")
plt.show()
```
